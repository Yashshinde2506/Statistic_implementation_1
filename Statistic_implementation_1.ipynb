```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# normal/gaussian dist
mu, sigma = 0, 0.1
s = np.random.normal(mu, sigma, size=1000)
print(len(s))
s
```

```python
plt.hist(s, 40, density=True)
plt.show()
```

```python
sns.distplot(s)
```

```python
# uniform dist >> all outcomes have equal prob
u = np.random.uniform(1,0,1000)
u
```

```python
plt.hist(u)
```

```python
# log normal dist
l = np.random.lognormal(mean=0, sigma=0.5, size=1000)
l
```

```python
plt.hist(l)
```

```python
sns.distplot(l)
```

```python
# log normal to log normal
sns.distplot(np.log(l))
```

```python
# sqrt transformation
sns.distplot(np.sqrt(l))
```

```python
# chisquare
c=np.random.chisquare(df=1, size=10000)
c = c[np.where(abs(c)<10)]
plt.hist(c, bins=100)
plt.show()
```

```python
c=np.random.chisquare(df=5, size=10000)
c = c[np.where(abs(c)<10)]
plt.hist(c, bins=100)
plt.show()
```

```python
# T dist, ss<30
t=np.random.standard_t(df=1, size=10000)
t = t[np.where(abs(t)<10)]
plt.hist(t, bins=100)
plt.show()
```

```python
# f dist
f=np.random.f(dfnum=1, dfden=1, size=10000)
f= f[np.where(abs(f)<10)]
plt.hist(f, bins=100)
plt.show()
```

```python
# 1. bernoulli dist

b = np.random.binomial(n=1, p=0.5, size=1000)
plt.hist(b)
plt.show()
```

```python
# 2. poisson dist
b=np.random.poisson(lam=5, size=1000)
plt.hist(b, bins=20)
plt.show()
```
